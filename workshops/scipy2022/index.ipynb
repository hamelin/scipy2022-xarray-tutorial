```python
import numpy as np
```

```python
np.arange(20).reshape((4,5))
```

```python
a = xr.DataArray(
    data=np.arange(20).reshape((4, 5)),
    coords={"x": [1,2,3,4], "y": [4,5,6,7,8]}
)
a
```

```python
b = xr.DataArray(
    data=np.arange(20).reshape((4, 5)),
    coords={"p": [1,2,3,4], "q": [4,5,6,7,8]}
)
b
```

```python
a * b
```

```python
a@b
```

```python
a @ b == a * b
```

```python
a * b
```

```python
(a * b) @ b
```
